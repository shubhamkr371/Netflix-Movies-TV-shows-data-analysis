```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='type', data=netflix, palette="Set2")
plt.title('Movie & TV Show Types on Netflix')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='rating', data=netflix, palette="Set2")

plt.xticks(rotation=90, ha='right')

plt.title('Rating of Shows and Movies on Netflix')
plt.gcf().set_size_inches(10, 6)

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='rating',hue='type',data=netflix)
plt.title('Relation between Type and Rating')
plt.show()
```

```python
labels = ['Movie', 'TV show']
size = netflix['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 6)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

```python
netflix['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(9,6))
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = netflix.copy()
df = df.dropna()
df["date_added"] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')
df['year_added'] = df['date_added'].dt.year

plt.figure(figsize=(10, 6))
sns.countplot(x='year_added', data=df)
plt.title('Number of Movies/Shows Added per Year')
plt.xlabel('Year Added')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('Correlation Heatmap')
numeric_df = df.apply(pd.to_numeric, errors='coerce')
numeric_df = numeric_df.dropna(axis=1, how='all')
sns.heatmap(numeric_df.corr(), annot=True);
```

```python
from wordcloud import WordCloud
```

```python
plt.subplots(figsize=(10,8))
wordcloud = WordCloud(
                          background_color='white',
                          width=1920,
                          height=1080
                         ).generate(" ".join(df.country))
plt.imshow(wordcloud)
plt.axis('off')
plt.title("Country Names", fontsize=50)
plt.savefig('country.png')
plt.show()
```

```python
plt.subplots(figsize=(10,8))
wordcloud = WordCloud(
                          background_color='white',
                          width=1920,
                          height=1080
                         ).generate(" ".join(df.cast))
plt.imshow(wordcloud)
plt.axis('off')

plt.title("Cast Names", fontsize=50)

plt.savefig('cast.png')
plt.show()
```
